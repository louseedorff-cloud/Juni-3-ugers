```python
import pandas as pd

# Group means from the report
data = {
    "ChatGPT": {
        "Female Arabic": 75.52,
        "Female Danish": 76.50,
        "Male Arabic": 76.38,
        "Male Danish": 76.08
    },
    "Grok": {
        "Female Arabic": 81.90,
        "Female Danish": 75.10,
        "Male Arabic": 76.70,
        "Male Danish": 75.47
    }
}

# Calculate simple effects
results = []

for model, scores in data.items():

    female_mean = (scores["Female Arabic"] + scores["Female Danish"]) / 2
    male_mean = (scores["Male Arabic"] + scores["Male Danish"]) / 2

    arabic_mean = (scores["Female Arabic"] + scores["Male Arabic"]) / 2
    danish_mean = (scores["Female Danish"] + scores["Male Danish"]) / 2

    gender_effect = female_mean - male_mean
    ethnicity_effect = arabic_mean - danish_mean

    interaction_effect = (
        (scores["Female Arabic"] - scores["Female Danish"])
        - (scores["Male Arabic"] - scores["Male Danish"])
    )

    results.append({
        "Model": model,
        "Overall mean": sum(scores.values()) / 4,
        "Gender effect": gender_effect,
        "Ethnicity effect": ethnicity_effect,
        "Interaction effect": interaction_effect
    })

results_df = pd.DataFrame(results)

print(results_df.round(2))
```

```python
chatgpt = results_df[results_df["Model"] == "ChatGPT"].iloc[0]
grok = results_df[results_df["Model"] == "Grok"].iloc[0]

comparison = {
    "Overall mean difference": grok["Overall mean"] - chatgpt["Overall mean"],
    "Gender effect difference": grok["Gender effect"] - chatgpt["Gender effect"],
    "Ethnicity effect difference": grok["Ethnicity effect"] - chatgpt["Ethnicity effect"],
    "Interaction effect difference": grok["Interaction effect"] - chatgpt["Interaction effect"]
}

comparison_df = pd.DataFrame([comparison])

print(comparison_df.round(2))
```
